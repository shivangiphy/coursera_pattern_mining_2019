```python
rows=[row.split(';') for row in data[0].values.tolist()]
```

```python
rows[0]
```

```python
single_itemset={}
for row in rows:
    for item in row:
#        print(item)
        if item not in single_itemset:
            single_itemset[item]=0
        single_itemset[item]+=1
        
#print(single_itemset)
        
        
    
    
```

```python
minsup=771
frequent_1_itemset=[]
for key,value in single_itemset.items():
    if value > minsup:
        frequent_1_itemset.append([value,key])

        
```

```python
df = pd.DataFrame(frequent_1_itemset)
```

```python
df.to_csv('patterns_part1.txt',sep=':',index=False, header=False)
```

```python
double_itemset={}
for row in rows:
    new_row=[]
    for item in row:
        if single_itemset[item]>minsup:
            new_row.append(item)
    for i, item1 in enumerate(new_row):
        for item2 in new_row[i+1:]:
            freq_2_item=tuple(sorted([item1,item2]))
        

            if freq_2_item not in double_itemset:
                double_itemset[freq_2_item]=0
            double_itemset[freq_2_item]+=1
        

    
```

```python
frequent_2_itemset=[]
for key,value in double_itemset.items():
    if value > minsup:
        frequent_2_itemset.append([value,';'.join(key)])
```

```python
len(frequent_2_itemset),len(frequent_1_itemset)
```

```python
triple_itemset={}
for row in rows:
    new_row=[]
    for item in row:
        if single_itemset[item]>minsup:
            new_row.append(item)
    for i, item1 in enumerate(new_row):
        for j, item2 in enumerate(new_row[i+1:],start=i+1):
            for item3 in new_row[j+1:]:
                freq_3_item=tuple(sorted([item1,item2,item3]))
        

                if freq_3_item not in triple_itemset:
                    triple_itemset[freq_3_item]=0
                triple_itemset[freq_3_item]+=1
        
frequent_3_itemset=[]
for key,value in triple_itemset.items():
    if value > minsup:
        frequent_3_itemset.append([value,';'.join(key)])
len(frequent_3_itemset)   
```

```python
four_itemset={}
for row in rows:
    new_row=[]
    for item in row:
        if single_itemset[item]>minsup:
            new_row.append(item)
    for i, item1 in enumerate(new_row):
        for j, item2 in enumerate(new_row[i+1:],start=i+1):
            for k, item3 in enumerate(new_row[j+1:],start=j+1):
                for item4 in new_row[k+1:]:
                    freq_4_item=tuple(sorted([item1,item2,item3,item4]))
        

                    if freq_4_item not in four_itemset:
                        four_itemset[freq_4_item]=0
                    four_itemset[freq_4_item]+=1
        
frequent_4_itemset=[]
for key,value in four_itemset.items():
    if value > minsup:
        frequent_4_itemset.append([value,';'.join(key)])
len(frequent_4_itemset)   
```
